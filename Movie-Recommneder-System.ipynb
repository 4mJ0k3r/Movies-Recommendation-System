```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv') 
```

```python
movies = movies.merge(credits,on='title') #merging the two dataframes on the basis of title

#Now taking only the columns that are required
# budget
# homepage
# id
# original_language
# original_title
# popularity
# production_comapny
# production_countries
# release-date(not sure)
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
movies.head()
```

```python
#This function extracts name values from the string tag - Genres,Keywords,Cast,Crew
import ast
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 

#Now dropping null values
movies.dropna(inplace=True)

#Applying the above function on the columns
movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
#This function extracts name values from the cast column but of the top 3 cast members
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L 

movies['cast'] = movies['cast'].apply(convert)
movies.head()
```

```python
movies['cast'] = movies['cast'].apply(lambda x:x[0:3]) #taking only the top 3 cast members
```

```python
#This function extracts name values from the crew column but of the director
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 

movies['crew'] = movies['crew'].apply(fetch_director)

movies.sample(5)
```

```python
#This function combines all the values of the columns into a single string and removes the spaces
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1

movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split()) #splitting the overview into words

movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew'] # Combining all the columns into a single list of tags
new = movies.drop(columns=['overview','genres','keywords','cast','crew']) #Dropping the old columns
new['tags'] = new['tags'].apply(lambda x: " ".join(x)) #Converting the list of tags into a single string
new.head()
```

```python
#Applying stemming to the tags
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)

new['tags'] = new['tags'].apply(stem) #(Here we removed duplicate words or similar words)
```

**This is the vectorisation of the tags**


```python
from sklearn.feature_extraction.text import CountVectorizer  #here we are converting the tags into vectors using the CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english') #max_features is the number of words to be considered and stop_words is the list of words to be ignored by the CountVectorizer 
    
vector = cv.fit_transform(new['tags']).toarray() #converting the tags into vectors
features = cv.get_feature_names_out()
print(features[50:90])
```

**Here we are finding the cosine similarity between the tags basically calculating the thetha angle between the two vectors**

```python

from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
print(similarity)
#new[new['title'] == 'The Lego Movie'].index[0]
def recommend(movie):  #This function recommends the top 5 movies based on the cosine similarity between the tags of the
    index = new[new['title'] == movie].index[0] #This is the index of the movie in the dataframe
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1]) #This sorts the movies based on the cosine similarity between the tags of the movie and the other movies in the dataframe , and lambda x: x[1] sorts the movies based on the cosine similarity in descending order
    for i in distances[1:6]:  #This prints the top 5 movies based on the cosine similarity between the tags of the movie
        print(new.iloc[i[0]].title)
        
    
recommend('Batman Begins')
```

**What if we want to save the dataframe and the similarity matrix into a pickle file?What is pickle?**

The pickle module in Python is used for serializing and storing objects into files. Serialization is the process of converting an object into a format that can be stored or transmitted, and in this case, the objects being serialized are the dataframe and the similarity matrix.<br> <br>
The pickle.dump() function is employed to serialize and store the dataframe and the similarity matrix into separate pickle files. The first argument specifies the object to be serialized, and the second argument is the file object obtained by opening the desired file in write-binary (wb) mode. This mode indicates that the data is being written in binary mode, and the objects are returned in a bytes object.<br> <br>
By using pickle.dump(), the dataframe and the similarity matrix are stored in binary format in the respective pickle files, allowing for easy retrieval and deserialization of the objects at a later time.

```python
import pickle

pickle.dump(new.to_dict(),open('movie_dict.pkl','wb')) #Here we are dumping the dataframe into a pickle file
pickle.dump(similarity,open('similarity.pkl','wb')) #Here we are dumping the similarity matrix into a pickle file
```
